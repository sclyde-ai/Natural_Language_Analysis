```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

絶対excelでcsvを開かないでください　データが壊れるかもしれない

```python
df = pd.read_csv('final_merged.csv', encoding='utf-8')
df
```

# 1800年と2019年の比較

```python
first_row = df.iloc[[0]]
last_row = df.iloc[[-1]]
diff_row = last_row.values - first_row.values
ratio_row = last_row.values / (first_row.values)
diff_df = pd.DataFrame(diff_row, columns=df.columns, index=['diff'])
ratio_df = pd.DataFrame(ratio_row, columns=df.columns, index=['ratio'])
result = pd.concat([first_row, last_row, diff_df, ratio_df])
result
```

```python
df_year_excluded = result.drop(columns=df.columns[0]) # without year
df_diff_ratio_only = df_year_excluded.iloc[[2, 3]]
df_diff_ratio_only
```

```python
row = df_diff_ratio_only.iloc[0]
max_val, max_col = row.max(), row.idxmax()
min_val, min_col = row.min(), row.idxmin()
max_val, max_col, min_val, min_col
```

```python
row = df_diff_ratio_only.iloc[0]

plt.bar(range(len(row)), row.values)
plt.xticks(range(len(row)), row.index)
plt.show()
```

```python
result_leq_negative = df_diff_ratio_only.loc[:, df_diff_ratio_only.iloc[0] <= -0.002]
result_leq_negative
```

```python
result_geq_positive = df_diff_ratio_only.loc[:, df_diff_ratio_only.iloc[0] >= 0.0005]
result_geq_positive
```

```python
result_very_stable = df_diff_ratio_only.loc[:, (df_diff_ratio_only.iloc[1] >= 0.95) & (df_diff_ratio_only.iloc[1] <= 1.05)]
result_very_stable
```

```python
result_significant = df_diff_ratio_only.loc[:, (df_diff_ratio_only.iloc[1] >= 10)]
result_significant
```

```python
result_very_significant = df_diff_ratio_only.loc[:, (df_diff_ratio_only.iloc[1] >= 100)]
result_very_significant
```

```python
average_diff = df_diff_ratio_only.iloc[0].mean()
df_without_inf = df.replace([np.inf, -np.inf], np.nan).dropna(axis=1, how='any')
average_ratio = df_without_inf.iloc[1].mean()
print("average_diff: " + str(average_diff) + "   " +"average_ratio: "+ str(average_ratio))
```
