```python
from google_pygram import GooglePyGram as gpg

def time_series_df(word_list, lang='English'):
    pygram = gpg(
        corpus=lang,
        # corpus_year=2019,
        # start_year=1800,
        # end_year=2000,
        # smoothing=3,
        case_sensitive=False,
        phrases=word_list
    )

    df = pygram.to_df().drop('year', axis=1)
    return df
```

```python
df = time_series_df(['hello', 'world'])

df.plot()
```
