```python
import gensim
from gensim.models import KeyedVectors

def load_and_use_chinese_model(model_path):
    
    print("Loading...")
    if model_path.endswith('.txt') or model_path.endswith('.bin'):
        wv = KeyedVectors.load_word2vec_format(model_path, binary=model_path.endswith('.bin'))
    elif model_path.endswith('.gensim') or model_path.endswith('.model'):
        wv = KeyedVectors.load(model_path)
    else:
        try:
            wv = KeyedVectors.load(model_path)
        except:
            wv = KeyedVectors.load_word2vec_format(model_path, binary=False)
    
    print(f"Loading Complete! Size of the table: {len(wv)}")
    
    test_words = ["中国", "人工智能", "北京", "上海"]
    for word in test_words:
        if word in wv:
            print(f"  '{word}' in the table")
        else:
            print(f"  '{word}' not found")
    
    return wv

# 使用
wv = load_and_use_chinese_model("tencent_vectors.gensim")
```

```python
vector = wv["中国"]
print(f"Vector Shape: {vector.shape}")
print(f"Vector type: {type(vector)}")
print(f"Top 10 values: {vector[:10]}")
```
