```python
from google_pygram import GooglePyGram as gpg
import pandas as pd
# グラフ描画のためにmatplotlibもインポートしておくと便利です
import matplotlib.pyplot as plt

# 検索したいフレーズをリストで指定
phrases = ['Albert Einstein', 'Sherlock Holmes', 'Frankenstein']
start_year = 1900
end_year = 2000
# コーパスを指定 (例: 'en-2019', 'en-2012', 'ja-2019'など)
corpus = 'en-2019'

try:
    # GooglePyGramのインスタンスを作成
    pygram = gpg(
        phrases=phrases,
        start_year=start_year,
        end_year=end_year,
        corpus=corpus,
        smoothing=3 # スムージング（平滑化）のレベル
    )

    # データをpandas DataFrameとして取得
    df = pygram.to_df()

    # 結果を表示
    print(df.head())

    # データをプロット
    df.plot(
        title=f'Google Ngram (Corpus: {corpus})',
        figsize=(10, 6)
    )
    plt.ylabel('Frequency')
    plt.grid(True)
    plt.show()

except Exception as e:
    print(f"エラーが発生しました: {e}")
    print("コーパス名が正しいか、ネットワーク接続を確認してください。")
```
