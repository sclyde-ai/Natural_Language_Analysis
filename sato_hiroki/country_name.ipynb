```python
import pycountry
import pandas as pd

print(pycountry.countries)

columns = ["alpha_2", "alpha_3", "name"]

English_matrix = []
for country in pycountry.countries:
    country_list = []
    country_list.append(country.alpha_2)
    country_list.append(country.alpha_3)
    country_list.append(country.name.replace(' ', '_'))
    English_matrix.append(country_list)
    # print(country)

Englis_df = pd.DataFrame(English_matrix, columns=columns)
Englis_df = Englis_df.set_index("alpha_2")
Englis_df

```

```python
from babel import Locale

locale = Locale.parse('zh_CN')

territories = locale.territories

country_dict = {}
for code, name in territories.items():
    if code.isalpha():
        country_dict[code] = name
        # print(code, name)

Chinese_series = pd.Series(country_dict, name="CN")
Chinese_series
```

```python
from babel import Locale

locale = Locale.parse('zh_TW')

territories = locale.territories

country_dict = {}
for code, name in territories.items():
    if code.isalpha():
        country_dict[code] = name
        # print(code, name)

Taiwan_series = pd.Series(country_dict, name="TW")
Taiwan_series
```

```python
df = Englis_df.drop(columns=["alpha_3"])
df.columns = ["EN"]
df = df.join(Chinese_series)
df = df.join(Taiwan_series)
df.to_csv("../data/country_names.csv")
df
```
